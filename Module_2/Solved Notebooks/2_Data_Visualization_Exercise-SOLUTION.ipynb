#### Visualize the total death for any 5 states (e.g. Total deaths for TX, CA, FL, PA, NJ)

```python
states = df[(df['state']== 'NY') | (df['state']== 'TX') | (df['state']== 'FL') | (df['state']== 'PA') | (df['state']== 'NJ')][['state','tot_death']]

ax = states.groupby('state').sum().plot(figsize=(20, 10), linewidth=2.5, kind = "bar", stacked = 'True', title = "Confirmed Cases in TX, FL, PA, ", grid ="true")
```

#### Visualize the percentage of confirmed cases for any 5 states. 

```python
states_conf = df[(df['state']== 'DE') | (df['state']== 'TX') | (df['state']== 'FL') | (df['state']== 'PA') | (df['state']== 'NJ')][['state','conf_cases']]

# states_conf.groupby('state').sum().plot(figsize=(10, 10), linewidth=2.5, kind = "bar", stacked = 'True', title = "Confirmed Cases in TX, FL, PA, ", grid ="true")

states_conf.groupby('state').sum().plot(figsize=(10, 10), kind = "pie", y = 'conf_cases', autopct='%1.0f%%', title='Confirmed Cases in TX, FL, PA')
```

#### Generate a corellation heathmap. Explain which features (columns) are highly corellated.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

corr=df.corr()
corr

plt.figure(figsize=(12,10))

plot = sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")
```
